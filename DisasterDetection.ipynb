## **3. Data Preprocessing**

### **3.1 Text Preprocessing**
Text preprocessing is a crucial step in Natural Language Processing. The following steps were applied to the `text` column to prepare it for modeling:

1. **Removing URLs**: Tweets often contain URLs, which are not relevant for the analysis and were thus removed.
2. **Removing Punctuation**: Punctuation marks were removed as they do not contribute to the semantic meaning of the text.
3. **Removing Stopwords**: Common stopwords such as "and," "the," etc., which do not carry significant meaning, were removed from the text.
4. **Stemming**: Words were reduced to their root forms using the Snowball Stemmer to standardize the vocabulary.

This preprocessing step was essential to ensure that the text data was clean and standardized, making it more suitable for modeling.

### **3.2 Balancing the Dataset**
Given the slight imbalance in the dataset, undersampling was applied to the majority class (target = 0) to balance the data. This step involved reducing the number of non-disaster-related tweets to match the number of disaster-related tweets, ensuring that the model would not be biased towards the majority class.

### **3.3 Splitting the Data**
The dataset was then split into training and validation sets. This split was necessary to evaluate the model's performance during the training phase and to ensure that the model could generalize well to unseen data.

### **3.4 Text Tokenization and Padding**
To prepare the text data for input into the neural network, the text was tokenized and the sequences were padded. Tokenization involved converting the text into sequences of integers, where each integer represented a word in the vocabulary. Padding was used to ensure that all sequences had the same length, which is a requirement for input into the neural network.

## **4. Model Architecture**

### **4.1 Model Selection**
Given the sequential nature of the text data, a Long Short-Term Memory (LSTM) model was chosen for this task. LSTM is a type of Recurrent Neural Network (RNN) that is well-suited for capturing dependencies in sequential data, making it a suitable choice for text classification tasks such as this one.

### **4.2 Word Embedding**
To convert the text data into a format that the LSTM model could process, word embeddings were used. The Keras embedding layer was employed to convert the integer sequences into dense vectors of fixed size, providing a rich representation of the text that the LSTM could use to learn patterns in the data.

### **4.3 Model Compilation**
The model was compiled using the Adam optimizer and a binary cross-entropy loss function. The Area Under the ROC Curve (AUC) metric was chosen to evaluate the model's performance, as it provides a robust measure of the model's ability to distinguish between the two classes.

### **4.4 Model Training**
The model was trained for 10 epochs with a batch size of 64. During training, the model was fitted on the training data and its performance was validated on the validation data. This training process involved adjusting the model's weights to minimize the loss function, while monitoring the AUC metric to ensure that the model was learning to distinguish between disaster-related and non-disaster-related tweets.

# Preprocessing
## Setting Up Tools for Text Preprocessing

In this step, we initialize essential tools for text preprocessing. We load the list of common English stopwords to filter out non-informative words, and we set up the Snowball Stemmer to reduce words to their base forms, helping to standardize the text data for better model performance.

```python
# Initialize the set of English stopwords and a stemmer for text preprocessing
stop_words = set(stopwords.words('english'))  # Load the list of common English stopwords
stemmer = SnowballStemmer("english")  # Initialize the Snowball stemmer for English
```

## Balancing the Dataset by Undersampling

To prevent the model from becoming biased towards the majority class (non-disaster tweets), we balance the dataset through undersampling. This process involves randomly reducing the number of non-disaster tweets to match the number of disaster-related tweets, ensuring that the model receives an equal representation of both classes during training.

```python
# Balancing the dataset by undersampling the majority class (non-disaster tweets)
class_0 = df_data[df_data['target'] == 0]  # Select all non-disaster-related tweets (target = 0)
class_1 = df_data[df_data['target'] == 1]  # Select all disaster-related tweets (target = 1)

# Undersample class 0 to match the size of class 1
class_0_under = class_0.sample(len(class_1), random_state=123)  # Randomly sample class 0 to the same size as class 1

# Combine the undersampled class 0 and class 1 to create a balanced dataset
df_data = pd.concat([class_0_under, class_1])[['id','text','target']]  # Keep only the 'id', 'text', and 'target' columns
```

## Shuffling and Splitting the Dataset

To ensure randomness in the distribution of examples, we shuffle the dataset before splitting it into training and validation sets. The training set, consisting of 80% of the data, is used to train the model, while the remaining 20% is reserved for validation, enabling us to evaluate the model's performance on unseen data.

```python
# Shuffle the dataset to ensure a random distribution of examples
df_data = shuffle(df_data)

# Split the dataset into training and validation sets
df_train, df_val = train_test_split(df_data, test_size=0.20)  # 80% of the data for training, 20% for validation
```

## Converting Text and Target Columns to NumPy Arrays

We convert the 'text' and 'target' columns from both the training and validation sets into NumPy arrays. This conversion facilitates efficient manipulation and processing of the data in subsequent steps, preparing it for model input and training.

```python
# Convert the 'text' and 'target' columns from the training and validation sets to NumPy arrays
text = df_train['text'].to_numpy()  # Training text data
label = df_train['target'].to_numpy()  # Training labels (targets)
text_val = df_val['text'].to_numpy()  # Validation text data
label_val = df_val['target'].to_numpy()  # Validation labels (targets)

text  # Display the training text data (as a NumPy array)
```

## Preprocessing Sentences for Text Cleaning

The `preprocess_sentence` function is designed to clean and prepare text data for modeling. It removes URLs, hashtags, and punctuation from the text, and filters out common stopwords, leaving only the most meaningful words. Finally, the cleaned words are joined back into a single string, ready for further processing or analysis.

```python
def preprocess_sentence(sentence):
    # Remove URLs from the text
    sentence = re.sub(r'http\S+|www\S+|https\S+', '', sentence, flags=re.MULTILINE)
    
    # Remove hashtags (only the # symbol, keep the text)
    sentence = re.sub(r'#\S+', '', sentence)
    
    # Remove punctuation from the text
    sentence = sentence.translate(str.maketrans('', '', string.punctuation))
    
    # Split the sentence into individual words
    words = sentence.split()
    
    # Remove stopwords from the text
    words = [word for word in words if word.lower() not in stop_words]
    
    # Join the words back into a single string
    return ' '.join(words)
```

## Preprocessing and Tokenizing the Text Data

In this section, we apply the preprocessing function to clean the training text data, followed by stemming each word to reduce it to its root form. After preprocessing, we initialize a tokenizer to convert the cleaned text into sequences of integers, which represent words. The tokenizer is then fitted on the text data to build a vocabulary, and the total vocabulary length is calculated, which will be used in the embedding layer of the neural network.

```python
# Apply the preprocessing function to each sentence in the training text data
text = [preprocess_sentence(x) for x in text]
print(text[0])  # Print the first preprocessed sentence to check the output

# Apply stemming to each word in the preprocessed text data
text = [stemmer.stem(word) for word in text]
print(text[0])  # Print the first stemmed sentence to check the output

# Initialize a tokenizer to convert words to sequences of integers
word_tokenizer = tf.keras.preprocessing.text.Tokenizer()

# Fit the tokenizer on the training text data to build the vocabulary
word_tokenizer.fit_on_texts(text)

# Calculate the vocabulary length (number of unique words)
vocab_length = len(word_tokenizer.word_index) + 1
```

## Comprehensive Text Preprocessing Function

The `preprocess_text` function applies several preprocessing steps to the text data. It first prints the original text for reference, then removes URLs, punctuation, and stopwords, followed by stemming the words to their root forms. After preprocessing, the function converts the text into sequences of integers and pads these sequences to a uniform length, ensuring they are ready for input into the neural network. Each major step is printed to allow for inspection and verification of the preprocessing process.#

```python
def preprocess_text(text):
    # Print the first original text for inspection
    print(text[0])
    
    # Preprocess each sentence in the text data (remove URLs, punctuation, stopwords)
    text = [preprocess_sentence(x) for x in text]
    
    # Print the first preprocessed sentence to verify the changes
    print(text[0])
    
    # Apply stemming to each word in the preprocessed text data
    text = [stemmer.stem(word) for word in text]
    
    # Print the first stemmed sentence to verify the stemming
    print(text[0])
    
    # Convert the text data into sequences of integers based on the tokenizer's vocabulary
    sequences = word_tokenizer.texts_to_sequences(text)
    
    # Print the first sequence of integers for inspection
    print(sequences[0])
    
    # Pad the sequences to ensure they all have the same length
    padded_sequences = pad_sequences(sequences)
    
    # Print the first padded sequence to verify the padding
    print(padded_sequences[0])
    
    # Return the padded sequences
    return padded_sequences
```

## Preparing Training and Validation Data for Model Input

In this step, we preprocess both the training and validation text data by converting the text into sequences of integers and padding them to ensure uniform input size. Additionally, the labels are converted into a one-hot encoded format, which is essential for training the neural network model in a binary classification task.

```python
# Preprocess the training text data (convert to sequences and pad them)
texts = preprocess_text(text)

# Convert the training labels to one-hot encoded format
labels = to_categorical(label)

# Preprocess the validation text data (convert to sequences and pad them)
texts_val = preprocess_text(text_val)

# Convert the validation labels to one-hot encoded format
labels_val = to_categorical(label_val)
```

## Observations on Preprocessing and Sequence Conversion

The output from the preprocessing and sequence conversion reveals that the original text has been successfully transformed into numerical sequences that represent the words based on the tokenizer's vocabulary. The padded sequences demonstrate that all sequences have been adjusted to a uniform length, ensuring compatibility with the input requirements of the neural network. Notably, the preprocessing steps, including the removal of stopwords and stemming, have effectively reduced the text to its most essential components, potentially improving the model's ability to generalize and focus on key information in the tweets. This process is crucial for maintaining consistency across the dataset and preparing it for effective learning during model training.

# Train

## Building and Training the LSTM Model

In this section, we construct a Sequential model designed to classify tweets as disaster-related or not. The model starts with an Embedding layer that transforms word indices into dense vectors, capturing semantic relationships between words. This is followed by an LSTM layer, which is particularly effective in handling sequential data by retaining information over time steps, making it ideal for processing text. A final Dense layer with a sigmoid activation function outputs probabilities for binary classification.

The model is compiled using the Adam optimizer and binary cross-entropy loss, with the AUC metric to evaluate performance during training. After defining the architecture, the model is trained on the preprocessed data over 10 epochs with a batch size of 64. During training, the model's performance is monitored on a separate validation set to ensure that it generalizes well to unseen data.

This setup is designed to capture the temporal dependencies within the text, which is crucial for understanding the context and meaning of disaster-related tweets. The use of dropout, though commented out here, could be explored further to mitigate overfitting by adding regularization during training.


```python
from keras.optimizers import Nadam

# Initialize a Sequential model
model = Sequential()

# Add an Embedding layer to convert word indices into dense vectors of fixed size (128-dimensional)
model.add(Embedding(vocab_length, 128))

# Optionally add a Dropout layer for regularization
model.add(Dropout(0.5))

# Add an LSTM layer with 64 units to capture sequential dependencies in the text data
model.add(LSTM(64))

# Add a Dense output layer with 2 units (for binary classification) and a sigmoid activation function
model.add(Dense(2, activation='sigmoid'))

# Compile the model with binary cross-entropy loss, Nadam optimizer, and AUC as the evaluation metric
model.compile(loss='BinaryCrossentropy', optimizer=Nadam(), metrics=['AUC'])

# Optionally, print the model summary to review the architecture
model.summary()

# Train the model on the preprocessed training data, validating on the validation data
# Run for 10 epochs with a batch size of 64
history = model.fit(texts, labels, validation_data=(texts_val, labels_val), epochs=10, batch_size=64)

# Plotting AUC and Loss for training and validation sets
plt.figure(figsize=(14, 6))

# Plot for AUC
plt.subplot(1, 2, 1)
plt.plot(history.history['AUC'], label='Train AUC')
plt.plot(history.history['val_AUC'], label='Validation AUC')
plt.title('AUC Over Epochs')
plt.xlabel('Epoch')
plt.ylabel('AUC')
plt.legend(loc='lower right')

# Plot for Loss
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Loss Over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(loc='upper right')

plt.tight_layout()
plt.show()
```

## Preparing and Saving Predictions for Submission

After generating the predictions, we extract the probabilities associated with the positive class (indicating disaster-related tweets). These probabilities are then converted into binary predictions based on a threshold of 0.5. The results are organized into a DataFrame that includes the tweet IDs from the test dataset and the corresponding binary predictions. Finally, this DataFrame is saved as a CSV file, ready for submission to the Kaggle competition.

```python
# Extract the probabilities for the positive class (class 1) from the predictions
prob_class_1 = predictions[:, 1]

# Convert probabilities to binary predictions (1 if probability >= 0.5, else 0)
binary_predictions = [1 if p >= 0.5 else 0 for p in prob_class_1]

# Create a DataFrame with the 'id' from the test dataset and the corresponding predictions
df = pd.DataFrame({'target': binary_predictions}, index=df_test['id'])

# Set the index name to 'id'
df.index.name = 'id'

# Save the predictions to a CSV file for submission
df.to_csv('/kaggle/working/submission.csv')
```
