```python
class Solution(object):
    def mergeAlternately(self, word1, word2):
        """
        :type word1: str
        :type word2: str
        :rtype: str
        """
        new_word = ""
        length1 = len(word1)
        length2 = len(word2)
        if length1 >= length2:
            for count in range(length2):
                new_word = new_word + word1[count]
                new_word = new_word + word2[count]
            if length1 > length2:
                new_word = new_word + word1[-(length1-length2):]
        else:
            for count in range(length1):
                new_word = new_word + word1[count]
                new_word = new_word + word2[count]
            new_word = new_word + word2[-(length2-length1):]
        return new_word
```

```python
word1 = ["abc", "ab", "abcd"]
word2 = ["pqr", "pqrs", "pq"]
n = 3
test = Solution()
for i in range(n):
    result = test.mergeAlternately(word1[i], word2[i])
    print(result)
```
