This project explores the application of Generative Adversarial Networks (GANs) to generate images in the style of Claude Monet. The GAN model consists of two main components: a generator and a discriminator. The generator's objective is to create images that mimic Monet's style, while the discriminator aims to distinguish between real Monet paintings and those generated by the model. The interaction between these two networks allows the generator to improve over time, ultimately producing images that closely resemble Monet's work.

This project leverages the dataset provided by Kaggle's "Monet Painting" competition, which includes both Monet paintings and photographs. The project focuses on transforming real-world photos into Monet-style artworks using the GAN framework.

## Model Architecture

The core of this project is the GAN model, which consists of a generator and a discriminator. The architecture of each model was designed to suit the task of transforming photos into Monet-style paintings.

### Generator

The generator model is responsible for creating Monet-style images from input photos. It consists of several convolutional layers for downsampling the image, followed by a series of upsampling layers to reconstruct the image in Monet's style.

#### Downsampling Layers

The downsampling layers use convolutional operations to extract features from the input image. These layers are followed by batch normalization and LeakyReLU activation functions to stabilize the training process and accelerate convergence.

#### Upsampling Layers

After downsampling, the image is passed through upsampling layers that use transpose convolution to increase the spatial dimensions of the feature maps. These layers also include batch normalization and LeakyReLU activation functions. The final output layer uses a `tanh` activation function to produce the output image.

### Discriminator

The discriminator model is designed to differentiate between real Monet paintings and those generated by the generator. It consists of several convolutional layers that progressively downsample the input image.

#### Convolutional Layers

The convolutional layers in the discriminator use LeakyReLU activation and dropout layers to prevent overfitting. The final output layer uses a sigmoid activation function to produce a binary classification output, indicating whether the input image is real or generated.

### GAN Model

The GAN model is created by combining the generator and discriminator models. The generator's output is fed into the discriminator, which evaluates the authenticity of the generated image. The generator is trained to fool the discriminator, while the discriminator is trained to correctly identify real and generated images.

## Training Process

### Detecting Duplicate Images

In any dataset, duplicate images can negatively impact the performance of machine learning models by introducing redundancy. Identifying and removing these duplicates is crucial for ensuring that the dataset is clean and that the model is trained on unique samples.

In the code below, we calculate the hash of each image file to identify duplicates. By comparing the hashes of all images within the Monet paintings and photos directories, we can detect which images are repeated. The number of duplicates found in each dataset is then printed out.

This process helps to maintain the quality of the dataset by eliminating unnecessary repetitions, which could otherwise lead to overfitting or skewed model performance.

### Preprocessing: Brightness Normalization and Augmentation

To ensure that the dataset is well-prepared for training the GAN model, we perform two key preprocessing steps: brightness normalization and brightness augmentation. These steps are crucial for enhancing the quality and diversity of the data, which in turn helps improve the model's performance.

1. **Brightness Normalization:** 
   - The images in the Monet paintings and photo datasets may have varying levels of brightness due to differences in lighting conditions and artistic style. Normalizing the brightness ensures that all images have a consistent brightness level, which helps prevent the model from being biased towards certain lighting conditions.

2. **Brightness Augmentation:**
   - After normalization, we apply brightness augmentation to introduce variability in the brightness levels of the images. This step is important for helping the model generalize better by learning to generate images under different lighting conditions, which is especially useful for creating realistic outputs.

The following code implements these preprocessing steps, preparing the unique Monet paintings and photos for the next stage of model training:

```python
import numpy as np
from PIL import Image, ImageEnhance
import os
import shutil
import random

# Function to normalize the brightness of an image
def normalize_brightness(image, target_brightness=128):
    """
    Normalizes the brightness of an image to a target value.

    :param image: Input image (PIL Image).
    :param target_brightness: Target brightness value (default 128).
    :return: Normalized image (PIL Image).
    """
    grayscale_img = image.convert('L')
    current_brightness = np.mean(np.array(grayscale_img))
    
    # Calculate the brightness enhancement factor
    brightness_factor = target_brightness / current_brightness
    
    # Apply brightness correction
    enhancer = ImageEnhance.Brightness(image)
    normalized_image = enhancer.enhance(brightness_factor)
    
    return normalized_image

# Function to apply brightness augmentation to an image
def augment_brightness(image, max_variation=30):
    """
    Applies brightness augmentation to an image with a specified variation.

    :param image: Input image (PIL Image).
    :param max_variation: Maximum variation in brightness (default 30).
    :return: Augmented image (PIL Image).
    """
    variation = random.uniform(-max_variation, max_variation)
    enhancer = ImageEnhance.Brightness(image)
    augmented_image = enhancer.enhance(1 + variation / 100)
    
    return augmented_image

# Create a directory for unique Monet paintings
unique_monet_path = "/kaggle/working/unique_monet"
os.makedirs(unique_monet_path, exist_ok=True)

# Copy all Monet paintings since no duplicates were found earlier
for file_name in monet_files:
    shutil.copy(os.path.join(monet_paintings_path, file_name), os.path.join(unique_monet_path, file_name))

print(f"Unique Monet paintings have been copied to {unique_monet_path}")

# Normalize brightness for unique Monet paintings
normalized_monet_images = [normalize_brightness(Image.open(os.path.join(unique_monet_path, f))) for f in os.listdir(unique_monet_path)]

# Normalize brightness for unique photos
normalized_photo_images = [normalize_brightness(Image.open(os.path.join(unique_photos_path, f))) for f in os.listdir(unique_photos_path)]

# Apply brightness augmentation to the normalized Monet images
augmented_monet_images = [augment_brightness(img) for img in normalized_monet_images]

# Apply brightness augmentation to the normalized photos
augmented_photo_images = [augment_brightness(img) for img in normalized_photo_images]
```

### Calculating and Visualizing the Mean Image

The mean image of a dataset can provide valuable insights into its overall structure, color distribution, and other visual characteristics. This is particularly useful for understanding how colors and textures are distributed across images, which can influence the model training process.

In this example, we will calculate and visualize the mean image for each of the datasets: Monet paintings and photos. This will help us see which colors and textures are most frequently represented in the images and identify any potential patterns or anomalies in the data.

Below is the code to perform this task:

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import os

def calculate_mean_image(image_list):
    # Initialize an array of zeros to store the sum of all images
    mean_image = np.zeros_like(image_list[0], dtype=np.float64)
    
    # Sum all the images
    for img in image_list:
        mean_image += img
    
    # Divide the sum by the number of images to get the mean
    mean_image /= len(image_list)
    
    # Convert the result back to uint8 format
    return mean_image.astype(np.uint8)

# Load all images from the Monet paintings directory
monet_images = [np.array(Image.open(os.path.join(monet_paintings_path, f))) for f in monet_files]

# Load all images from the photos directory
photo_images = [np.array(Image.open(os.path.join(photos_path, f))) for f in photos_files]

# Calculate the mean image for Monet paintings
mean_monet = calculate_mean_image(monet_images)

# Calculate the mean image for photos
mean_photo = calculate_mean_image(photo_images)

# Visualize the mean image for Monet paintings
plt.imshow(mean_monet)
plt.title("Mean Monet Image")
plt.axis('off')
plt.show()

# Visualize the mean image for photos
plt.imshow(mean_photo)
plt.title("Mean Photo Image")
plt.axis('off')
plt.show()
```

By ensuring that the images are standardized and correctly formatted, we help the GAN learn and generate images more effectively.
```python
# Retrieve the list of Monet TFRecord files
MONETFILENAMES = [os.path.join(monet_tfrec_path, f) for f in os.listdir(monet_tfrec_path)]
print('Number of Monet TFRecord Files:', len(MONETFILENAMES))
print(MONETFILENAMES)

# Retrieve the list of photo TFRecord files
PHOTOFILENAMES = [os.path.join(photos_tfrec_path, f) for f in os.listdir(photos_tfrec_path)]
print('Number of Photo TFRecord Files:', len(PHOTOFILENAMES))
```

The following preprocessing code is adapted from the previously referenced example notebook. It handles the loading of .tfrec files containing the images and normalizes the pixel values to a range of -1 to 1, which is standard practice for training GANs. Additionally, the code ensures that each image has three color channels (RGB), making them ready for input into the model.

```python
# Define image dimensions and enable AUTOTUNE for optimized data loading
image_dimensions = [256, 256]  # width, height
AUTOTUNE = tf.data.experimental.AUTOTUNE

# Function to decode and preprocess the image
def decode_image(image):
    image = tf.image.decode_jpeg(image, channels=3)  # Ensure the image has 3 color channels (RGB)
    image = (tf.cast(image, tf.float32) / 127.5) - 1  # Normalize pixel values to the range [-1, 1]
    image = tf.reshape(image, [*image_dimensions, 3])  # Reshape image to the desired dimensions
    return image

# Function to parse the TFRecord file and extract image data
def read_tfrecord(example):
    tfrecord_format = {
        "image_name": tf.io.FixedLenFeature([], tf.string),
        "image": tf.io.FixedLenFeature([], tf.string),
        "target": tf.io.FixedLenFeature([], tf.string)
    }
    example = tf.io.parse_single_example(example, tfrecord_format)
    image = decode_image(example['image'])  # Decode the image using the decode_image function
    return image

# Function to load and preprocess the dataset from TFRecord files
def load_dataset(filenames, labeled=True, ordered=False):
    dataset = tf.data.TFRecordDataset(filenames)  # Load TFRecord files
    dataset = dataset.map(read_tfrecord, num_parallel_calls=AUTOTUNE)  # Decode images in parallel
    return dataset.prefetch(AUTOTUNE)  # Prefetch data for optimized loading

# Load and batch the Monet paintings and photos datasets
monet_ds = load_dataset(MONET_FILENAMES, labeled=True).batch(4).prefetch(AUTOTUNE)
photo_ds = load_dataset(PHOTO_FILENAMES, labeled=True).batch(4).prefetch(AUTOTUNE)

# Print the dataset objects to verify successful loading
print(monet_ds)
print(photo_ds)
```

# Model Architecture

In this GAN project, we employ two key components: the generator and the discriminator.

### Generator

The generator is designed to transform input images into the style of Monet’s paintings. It begins by taking in the image dimensions as input. The model starts with downsampling layers that apply convolution operations to extract essential features from the image. These layers are followed by batch normalization, which stabilizes the learning process and accelerates training by normalizing the activations. The LeakyReLU activation function is used to introduce non-linearity, which helps the network learn more complex patterns. After downsampling, the model proceeds with upsampling layers. These layers use convolution transpose operations to increase the spatial dimensions of the feature maps, essentially reconstructing the image with the target style. Again, batch normalization and LeakyReLU are applied to maintain stability and enable the network to learn effectively.

### Discriminator

The discriminator's role is to evaluate the authenticity of the images, distinguishing between real Monet paintings and those generated by the generator. It accomplishes this by downsampling the input images through a series of convolutional layers. LeakyReLU is used as the activation function, helping the network to learn without being affected by the problem of dying ReLUs. Dropout layers are included to prevent overfitting, ensuring that the discriminator does not become too confident or biased toward the training data. By alternating between training the generator and the discriminator, the GAN iteratively improves, with the generator creating increasingly realistic images and the discriminator becoming more adept at identifying fakes.

```python
# Code was inspired by the example notebook, with some modifications

# Generator
# Input layer
generator_input = Input(shape=(256, 256, 3))

# Downsample
g1 = Conv2D(64, 3, strides=2, padding='same')(generator_input)
g1 = LeakyReLU()(g1)

g2 = Conv2D(128, 3, strides=2, padding='same')(g1)
g2 = BatchNormalization()(g2)
g2 = LeakyReLU()(g2)

g3 = Conv2D(256, 3, strides=2, padding='same')(g2)
g3 = BatchNormalization()(g3)
g3 = LeakyReLU()(g3)

# Upsample
g4 = Conv2DTranspose(128, 3, strides=2, padding='same')(g3)
g4 = BatchNormalization()(g4)
g4 = LeakyReLU()(g4)
g4 = Concatenate()([g4, g2])

g5 = Conv2DTranspose(64, 3, strides=2, padding='same')(g4)
g5 = BatchNormalization()(g5)
g5 = LeakyReLU()(g5)
g5 = Concatenate()([g5, g1])

# Output layer
generator_output = Conv2DTranspose(3, 3, strides=2, padding='same', activation='tanh')(g5)

# Define the model
generator_model = Model(generator_input, generator_output)
generator_model.summary()


# Discriminator
# Input layer
discriminator_input = Input(shape=(256, 256, 3))

# Convolutional layers
d1 = Conv2D(64, 3, strides=2, padding='same')(discriminator_input)
d1 = LeakyReLU()(d1)
d1 = Dropout(0.3)(d1)

d2 = Conv2D(128, 3, strides=2, padding='same')(d1)
d2 = LeakyReLU()(d2)
d2 = Dropout(0.3)(d2)

d3 = Conv2D(256, 3, strides=2, padding='same')(d2)
d3 = LeakyReLU()(d3)
d3 = Dropout(0.3)(d3)

# Flatten and output layer
d4 = Flatten()(d3)
discriminator_output = Dense(1, activation='sigmoid')(d4)

# Define the model
discriminator_model = Model(discriminator_input, discriminator_output)

discriminator_model.summary()

# Compile the discriminator model
discriminator_model.compile(loss='binary_crossentropy', optimizer=Adam(0.001, 0.5), metrics=['accuracy'])
```

```python
# Define the GAN model by combining the generator and discriminator models
gan_model = Sequential([
    generator_model,
    discriminator_model
])

# Compile the GAN model with binary crossentropy loss and Adam optimizer
gan_model.compile(loss='binary_crossentropy', optimizer=Adam(0.001, 0.5))

# Helper function to plot the training history of both the discriminator and generator
def plot_training_history(discriminator_history, generator_history):
    # Unpack the loss and accuracy for both models
    d_loss, d_accuracy = zip(*discriminator_history)
    g_loss, g_accuracy = zip(*generator_history)
    
    epochs = range(1, len(d_loss) + 1)
    
    # Create subplots for loss and accuracy
    fig, ax = plt.subplots(2, 1, figsize=(10, 10))
    
    # Plot Discriminator Loss and Accuracy
    ax[0].plot(epochs, d_loss, 'bo-', label='Discriminator Loss')
    ax[0].plot(epochs, d_accuracy, 'ro-', label='Discriminator Accuracy')
    ax[0].set_title('Discriminator Loss and Accuracy')
    ax[0].set_xlabel('Epochs')
    ax[0].set_ylabel('Loss/Accuracy')
    ax[0].legend()
    
    # Plot Generator Loss and Accuracy
    ax[1].plot(epochs, g_loss, 'bo-', label='Generator Loss')
    ax[1].plot(epochs, g_accuracy, 'ro-', label='Generator Accuracy')
    ax[1].set_title('Generator Loss and Accuracy')
    ax[1].set_xlabel('Epochs')
    ax[1].set_ylabel('Loss/Accuracy')
    ax[1].legend()
    
    # Display the plots
    plt.tight_layout()
    plt.show()

# Helper function to display images generated by the generator model
def show_imgs(generator_model, real_photos):
    transformed_images = generator_model.predict(real_photos)  # Generate Monet-style images from real photos

    # Determine the grid size for displaying images
    num_images = transformed_images.shape[0]
    num_cols = int(np.sqrt(num_images))
    num_rows = num_cols
    
    # Plot the images in a grid
    fig, axs = plt.subplots(num_rows, num_cols, figsize=(5, 5))
    cnt = 0
    for i in range(num_rows):
        for j in range(num_cols):
            if cnt < num_images:
                axs[i, j].imshow((transformed_images[cnt] * 0.5 + 0.5))  # Rescale images to [0, 1] range for display
                axs[i, j].axis('off')
                cnt += 1
    plt.tight_layout()
    plt.show()
    plt.close()

# Function to train the GAN model
def train_gan(generator_model, discriminator_model, gan_model, photo_ds, monet_ds, epochs, batch_size):
    # Create labels for real and fake images
    real = np.ones((batch_size, 1))
    fake = np.zeros((batch_size, 1))

    # Lists to store the training history
    discriminator_losses = []
    generator_losses = []

    # Determine the number of steps per epoch based on dataset size
    total_photos = len(list(photo_ds))
    total_monets = len(list(monet_ds))
    steps_per_epoch = min(total_photos, total_monets) // batch_size

    for epoch in range(epochs):
        # Create iterators for the datasets
        photo_iter = iter(photo_ds.repeat())
        monet_iter = iter(monet_ds.repeat())

        # Iterate through the dataset for the current epoch
        for i in range(steps_per_epoch):
            # Load the next batch of real photos and Monet paintings
            real_photos = next(photo_iter)
            real_monets = next(monet_iter)

            # Generate fake Monet images from real photos
            fake_monets = generator_model.predict(real_photos, verbose=0)

            # Train the discriminator on real and fake images
            discriminator_model.trainable = True
            d_loss_real = discriminator_model.train_on_batch(real_monets, real)
            d_loss_fake = discriminator_model.train_on_batch(fake_monets, fake)
            d_loss = 0.5 * np.add(d_loss_real, d_loss_fake)  # Average discriminator loss
            discriminator_losses.append(d_loss)

            # Train the generator
            discriminator_model.trainable = False
            g_loss = gan_model.train_on_batch(real_photos, real)
            generator_losses.append(g_loss[1:])

            # Display training progress and generated images at intervals
            if (i) % 25 == 0:  # Adjust the frequency as needed
                print(f"Step {i+1}/{steps_per_epoch}, Epoch {epoch+1}/{epochs} - Discriminator Loss: {d_loss[0]}, Accuracy: {round(100 * d_loss[1], 2)}% - Generator Loss: {g_loss[1]}, Accuracy: {round(100 * g_loss[2], 2)}%")
                real_photos = next(iter(photo_ds.take(1)))  # Take a batch for display
                show_imgs(generator_model, real_photos)

    return discriminator_losses, generator_losses

# Train the GAN and plot the training history
discriminator_losses, generator_losses = train_gan(generator_model, discriminator_model, gan_model, photo_ds, monet_ds, epochs=25, batch_size=4)
plot_training_history(discriminator_losses, generator_losses)
```
